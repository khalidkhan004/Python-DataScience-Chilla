 
## EDA (Exploratory Data Analysis): 
Exploratory data analysis (EDA) is a method used to `analyze` and `summarize` the dataset.in pythonwe can used the pandas library to perform EDA on dataset 

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
df=sns.load_dataset("titanic")
```

```python
# show the first 5 records of dataset
df.head()
```

```python
# show the last 5 records of dataset
df.tail()
```

```python
# show the informations of a dataset using Info() functions
df.info()
```

```python
# shows the rows and columns of dataset
df.shape
```

```python
# shows the datatypes of dataset
df.dtypes
```

### start summarize


```python
# summarize (summary statistics)
# shows the mean ,std ,min ,max ,count using describe() functions
df.describe()
```

```python
# find the null values in the form of boolean (true , false) True = null ,false= fill
df.isnull()
```

```python
# to find the null values in the form of count
df.isnull().sum()
```

```python
# to find the null values in the form of percentage
df.isnull().sum() /len(df) *100
```

```python
# finding unique value of all dataset
df.nunique()
```

```python
# shows the columns
df.columns
```

```python
# visualize the nan value in heatmap
sns.heatmap(df.isnull() ,yticklabels=False ,cbar=False ,cmap='viridis')
```

```python
# remove columns
df.drop(columns="deck" ,inplace=True)
```

```python
# again check the columns
df.shape
```

```python
# visualize and explore
sns.countplot(data=df , x="survived")
```

```python
sns.barplot(data=df ,x="survived" ,y="age" ,hue="who")
```

```python
sns.countplot(x="pclass" , data=df)
```

```python
# data distributions
sns.histplot(x="age" ,data=df ,bins=30)
```

```python
df['age'].hist(bins=30 ,color='green' ,alpha=0.7)
```

```python
sns.boxplot(data=df ,x='survived' ,y='age' ,hue='who')
```

```python
sns.scatterplot(data=df ,x='age',y='fare',hue='pclass')
plt.ylim(0,300)
```

```python
sns.pairplot(df ,hue='survived')
```

```python
# stacked bar plot
df.groupby(['pclass' ,'who'])['fare'].mean().unstack().plot(kind='bar' ,stacked=True)
```

```python
sns.swarmplot(data=df ,x='survived' ,y='age', hue='who')
```

```python
sns.catplot(x='pclass' ,y='fare' ,hue='embarked',kind='point',data=df)
```

```python
sns.lmplot(data=df ,x='age',y='fare' ,hue='pclass')
plt.ylim(0 ,300)
```
